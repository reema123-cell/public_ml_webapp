# Importing Dataset

```python
dataset = pd.read_csv('heart.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

# Data Preprocessing

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder() 
le2 = LabelEncoder() 
le6 = LabelEncoder() 
le8 = LabelEncoder() 
le10 = LabelEncoder() 
x[:,1] = le1.fit_transform(x[:,1])
x[:,2] = le2.fit_transform(x[:,2])
x[:,6] = le6.fit_transform(x[:,6])
x[:,8] = le8.fit_transform(x[:,8])
x[:,10] = le10.fit_transform(x[:,10])
```

```python
print(x)
```

```python
print(y)
```

# Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_test)
```

# Training Dataset

##Training with Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model_randomforest = RandomForestClassifier()
model_randomforest.fit(X_train,Y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression()
model_logistic.fit(X_train,Y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_kneighbors = KNeighborsClassifier()
model_kneighbors.fit(X_train,Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_decision = DecisionTreeClassifier()
model_decision.fit(X_train,Y_train)
```

```python
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(X_train,Y_train)
```

```python
import joblib

# Load the saved scaler
scaler = joblib.load('scaler.pkl')




```

```python

# Load models and scaler
model_randomforest = joblib.load('random_forest_model.pkl')
model_svm = joblib.load('svm_model.pkl')
model_kneighbors = joblib.load('knn_model.pkl')
model_decision = joblib.load('decision_tree_model.pkl')
model_logistic = joblib.load('logistic_model.pkl')
scaler = joblib.load('scaler.pkl')

# Title of the app
st.title("Heart Disease Prediction App")

# Dropdown to select the model
model_choice = st.selectbox("Choose the model", options=["Random Forest", "SVM", "KNN", "Decision Tree", "Logistic Regression"])

# Input fields for health data
age = st.number_input("Age", min_value=1, max_value=120, value=30)
sex = st.selectbox("Sex", options=[0, 1])  # 0 = female, 1 = male
cp = st.selectbox("Chest Pain Type", options=[0, 1, 2, 3])  # Example options
trestbps = st.number_input("Resting Blood Pressure", min_value=0, max_value=300, value=120)
chol = st.number_input("Cholestoral", min_value=0, max_value=600, value=200)
fbs = st.selectbox("Fasting Blood Sugar > 120 mg/dl", options=[0, 1])  # 0 = No, 1 = Yes
restecg = st.selectbox("Resting Electrocardiographic Results", options=[0, 1, 2])  # Example options
thalach = st.number_input("Maximum Heart Rate Achieved", min_value=0, max_value=250, value=150)
exang = st.selectbox("Exercise Angina", options=[0, 1])  # 0 = No, 1 = Yes
oldpeak = st.number_input("Depression Induced by Exercise", min_value=0.0, max_value=10.0, value=0.0)
slope = st.selectbox("Slope of Peak Exercise ST Segment", options=[0, 1, 2])  # Example options


# Create an array of the input data
input_data = np.array([[
    age, sex, cp, trestbps, chol,fbs,restecg,thalach,exang,oldpeak,slope
]])

# Scale the input data (if required by your model)
input_data_scaled = sc.transform(input_data)

# Button to make prediction
if st.button("Predict Heart Disease"):
    if model_choice == "Random Forest":
        prediction = model_randomforest.predict(input_data_scaled)
    elif model_choice == "SVM":
        prediction = model_svm.predict(input_data_scaled)
    elif model_choice == "KNN":
        prediction = model_kneighbors.predict(input_data_scaled)
    elif model_choice == "Decision Tree":
        prediction = model_decision.predict(input_data_scaled)
    elif model_choice == "Logistic Regression":
        prediction = model_logistic.predict(input_data_scaled)
    
    # Output result
    if prediction == 1:
        st.write("The person has heart disease.")
    else:
        st.write("The person does not have heart disease.")
```

# Making Confusion Matrix

```python
y_pred_logistic = model_logistic.predict(X_test)
y_pred_neighbors = model_kneighbors.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_decision = model_decision.predict(X_test)
y_pred_random = model_randomforest.predict(X_test)
```

```python
RandomForest_Accuracy = accuracy_score(Y_test,y_pred_random)
LogisticRegression_Accuracy = accuracy_score(Y_test,y_pred_logistic)
KNeighbors_Accuracy = accuracy_score(Y_test,y_pred_neighbors)
SVM_Accuracy = accuracy_score(Y_test,y_pred_svm)
Decision_Accuracy = accuracy_score(Y_test,y_pred_decision)
```

```python
plt.figure(figsize=(12, 6))
plt.bar("Logistic Regression",LogisticRegression_Accuracy,width=0.6)
plt.bar("KNeighbors",KNeighbors_Accuracy,width=0.6)
plt.bar("Support Vector Machine",SVM_Accuracy,width=0.6)
plt.bar("Decision Tree",Decision_Accuracy,width=0.6)
plt.bar("Random Forest",RandomForest_Accuracy,width=0.6)
plt.xlabel("Machine Learning Algorithm")
plt.ylabel("Accuracy")
```
